```python
import matplotlib.pyplot as plt
import os
```

```python
def parse_output(output):
    x_values = []
    y_values = []

    for line in output.split('\n'):
        if line.strip() != '':
            x, y = map(float, line.split())
            x_values.append(x)
            y_values.append(y)

    return x_values, y_values
```

```python
os.system("mpicc ./logistic.c")
```

```python
output = os.popen("mpirun -n 6 ./a.out").read()
r, attractors = parse_output(output)
```

```python
plt.plot(r, attractors)
plt.ylim(0, 17)
plt.xlim(2.8, 4.0)
plt.xlabel("r")
plt.ylabel("attractors")
plt.grid(True)
```
